```python
summary = data2.describe().T

summary["median"] = data2.median()

summary["variance"] = data2.var()

summary["skewness"] = data2.skew()

summary["kurtosis"] = data2.kurt()

summary
```

```python
for col in data2.columns[1:]:
    plt.figure(figsize=(6,3))
    sns.violinplot(x=data2[col])
    plt.title(col)
    plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(data2["Year"], data2["Total_Budget_Billions_USD"])
plt.title("Total Budget Over Years")
plt.xlabel("Year")
plt.ylabel("Budget (Billion USD)")
plt.grid(True)
plt.show()
```
