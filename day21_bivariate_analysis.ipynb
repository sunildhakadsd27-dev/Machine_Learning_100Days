```python
df['Budget_Growth_%'] = df['Total_Budget_Billions_USD'].pct_change()*100

plt.figure(figsize=(14,5))
sns.lineplot(data=df, x='Year', y='Budget_Growth_%')
plt.title("Year-on-Year Budget Growth")
plt.grid(True)
plt.show()
```

```python
start = df['Total_Budget_Billions_USD'].iloc[0]
end = df['Total_Budget_Billions_USD'].iloc[-1]
years = len(df)-1

cagr = ((end/start)**(1/years)-1)*100

print(f"CAGR = {cagr:.2f}%")
```

```python
df['Rolling_5Y'] = df['Total_Budget_Billions_USD'].rolling(5).mean()

plt.figure(figsize=(14,5))
plt.plot(df['Year'], df['Total_Budget_Billions_USD'], label='Actual')
plt.plot(df['Year'], df['Rolling_5Y'], label='5-Year Average')
plt.legend()
plt.show()
```

```python
cols = [
'Defense_Amount_Billions_USD',
'Education_Amount_Billions_USD',
'Health_Amount_Billions_USD',
'Infrastructure_Amount_Billions_USD',
'Agriculture_Amount_Billions_USD'
]

share = df[cols].div(df['Total_Budget_Billions_USD'], axis=0)*100
share.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df[[
'Defense_Amount_Billions_USD',
'Education_Amount_Billions_USD',
'Health_Amount_Billions_USD',
'Infrastructure_Amount_Billions_USD',
'Agriculture_Amount_Billions_USD'
]]

y = df['Total_Budget_Billions_USD']

model = RandomForestRegressor(random_state=42)
model.fit(X, y)

importance = pd.Series(model.feature_importances_, index=X.columns)
importance.sort_values().plot(kind='barh')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X_scaled = StandardScaler().fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

print(pca.explained_variance_ratio_)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)

df['Cluster'] = kmeans.fit_predict(X)
```

```python
sns.scatterplot(
data=df,
x='Defense_Amount_Billions_USD',
y='Education_Amount_Billions_USD',
hue='Cluster',
palette='Set2'
)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Year']]
y = df['Total_Budget_Billions_USD']

model = LinearRegression()
model.fit(X, y)

future = pd.DataFrame({'Year':[2027,2028,2029,2030]})

future['Prediction'] = model.predict(future)

print(future)
```

```python
Q1 = df['Total_Budget_Billions_USD'].quantile(0.25)
Q3 = df['Total_Budget_Billions_USD'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
(df['Total_Budget_Billions_USD'] < Q1 - 1.5*IQR) |
(df['Total_Budget_Billions_USD'] > Q3 + 1.5*IQR)
]

outliers
```
