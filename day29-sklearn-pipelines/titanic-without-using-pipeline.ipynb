```python
data.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.drop(columns="Survived"),data['Survived'],test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer

age_si = SimpleImputer()
embarked_si = SimpleImputer(strategy='most_frequent')

x_train_age_si = age_si.fit_transform(X_train[['Age']])
x_test_age_si = age_si.fit_transform(X_test[['Age']])

x_train_embarked_si = embarked_si.fit_transform(X_train[['Embarked']])
x_test_embarked_si = embarked_si.fit_transform(X_test[['Embarked']])
```

```python
x_test_age_si.shape
x_train_age_si.shape
```

```python
x_train_embarked_si.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(x_train_embarked_si)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(x_test_embarked_si)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train.head()
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,x_train_age_si,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,x_test_age_si,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train_transformed,y_train)
```

```python
y_pred = dtc.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
